````python
### crie a função ordena_por_comprimento e retorne a lista resultado
# Função que ordena uma lista de strings pelo comprimento
def ordena_por_comprimento(lista):
    return sorted(lista, key=lambda s: len(s))

# Exemplo de uso
strings = ["banana", "kiwi", "abacaxi", "manga", "maçã"]
strings_ordenadas = ordena_por_comprimento(strings)

print(strings_ordenadas)


### aplique sua função à seguinte lista e imprima o resultado
nomes = ["Joao", "Maria", "Jose", "Gabriela", "Sol", "Luna", "Bento", "Enzo", "Fernanda"]

````

### Q4.

Crie uma lista de listas (uma matriz de 2 dimensões) de tamanho $n \times n$, com $n$ dado pelo usuário. Cada elemento da matriz deve ser o produto dos índices da linha pela coluna.

Exemplo:
```
Digite n: 4
Matriz:
[[0, 0, 0, 0],
 [0, 1, 2, 3],
 [0, 2, 4, 6],
 [0, 3, 6, 9]]
```

````python
# Função para criar a matriz
def cria_matriz(n):
    return [[i * j for j in range(n)] for i in range(n)]

# Solicita ao usuário o tamanho da matriz
n = int(input("Digite o valor de n para uma matriz n x n: "))

# Cria a matriz
matriz = cria_matriz(n)

# Exibe a matriz
for linha in matriz:
    print(linha)
````

### Q5.

Dada uma lista de números inteiros, implemente uma função em Python chamada ```pares_unicos``` que encontre e retorne todos os pares únicos de elementos na lista cuja soma seja igual a um determinado valor. Considere que a ordem dos pares não é relevante.

````python

def pares_unicos(lista, soma_desejada):
    # Usaremos um conjunto para armazenar os pares únicos
    pares = set()
    
    # Conjunto para armazenar os elementos já vistos
    vistos = set()
    
    for num in lista:
        # Calcula o número necessário para completar a soma desejada
        complemento = soma_desejada - num
        
        # Verifica se o complemento já foi visto e forma um par
        if complemento in vistos:
            # Adiciona o par ordenado para garantir a unicidade
            pares.add(tuple(sorted((num, complemento))))
        
        # Adiciona o número atual ao conjunto de vistos
        vistos.add(num)
    
    # Retorna a lista de pares únicos como uma lista
    return list(pares)

# Exemplo de uso
lista = [1, 3, 2, 4, 3, 0, 5]
soma_desejada = 5
resultado = pares_unicos(lista, soma_desejada)

print(resultado)


nums = [3, 4, 5, 6, 7]
soma = 10
resultado = pares_unicos(nums, soma)
print(resultado)  # Saída esperada: [(3, 7), (4, 6)]
````

### Q5.

Escreva um programa para encontrar a diferença entre duas listas, incluindo elementos duplicados.

Ex:
```
Listas originais: 
[1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
[1, 1, 2, 4, 5, 6]
Diferença entre as listas:
[3, 3, 4, 7]
```

````python
from collections import Counter

def diferença_listas(lista1, lista2):
    # Conta a frequência de cada elemento nas listas
    counter1 = Counter(lista1)
    counter2 = Counter(lista2)
    
    # Calcula a diferença entre as duas contagens
    diferença = counter1 - counter2
    
    # Converte o resultado para uma lista
    resultado = list(diferença.elements())
    
    return resultado

# Exemplo de uso
lista1 = [1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
lista2 = [1, 1, 2, 4, 5, 6]

resultado = diferença_listas(lista1, lista2)
print("Diferença entre as listas:", resultado)
````

Dada uma lista aleatória, escreva um programa que separa os valores em listas menores de acordo com o tamanho solicitado. Caso não seja possível repartir igualmente a lista original, a última sublista deve conter os elementos remanescentes, como indicado no exemplo a seguir.

Note que seu programa deve continuar executando em um laço infinito até que o usuário informe o tamanho 0 na entrada solicitada.

```
Lista original:
[12, 45, 23, 67, 78, 90, 45, 32, 100, 76, 38, 62, 73, 29, 83]

Tamanho para divisão: 3
Subslistas:
[[12, 45, 23], [67, 78, 90], [45, 32, 100], [76, 38, 62], [73, 29, 83]]

Tamanho para divisão: 4
Subslistas: [[12, 45, 23, 67], [78, 90, 45, 32], [100, 76, 38, 62], [73, 29, 83]]

Tamanho para divisão: 0
```

````python
import random
lista = [random.randint(1,100) for i in range(20)]

def dividir_lista(lista, tamanho):
    # Lista para armazenar as sublistas
    sublistas = []
    
    # Cria as sublistas de acordo com o tamanho
    for i in range(0, len(lista), tamanho):
        sublistas.append(lista[i:i+tamanho])
    
    return sublistas

# Lista original
lista_original = [12, 45, 23, 67, 78, 90, 45, 32, 100, 76, 38, 62, 73, 29, 83]

# Laço infinito até que o tamanho 0 seja inserido
while True:
    # Solicita o tamanho para divisão
    tamanho = int(input("Tamanho para divisão: "))
    
    # Se o tamanho for 0, o programa encerra
    if tamanho == 0:
        break
    
    # Chama a função para dividir a lista
    sublistas = dividir_lista(lista_original, tamanho)
    
    # Exibe as sublistas
    print("Subslistas:", sublistas)
````
